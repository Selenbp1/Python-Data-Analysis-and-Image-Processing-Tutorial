```python
#cv2.findContours(image, mode, method):  Contour  

#mode: Contour  
#1) RETR_EXTERNAL:  Line 
#2) RETR_LIST:  Line , Hierarchy  X
#3) RETR_TREE:  Line ,  Hierarchy  O

#method: Contour   
#1) CHAIN_APPROX_NONE:  Contour  
#2) CHAIN_APPROX_SIMPLE: Contour Line     

#  Gray Scale Threshold   .
#cv2.drawContours(image, contours, contour_index, color, thickness): Contour  
#contour_index:   Contours Line (: -1)

import cv2
import matplotlib.pyplot as plt

image = cv2.imread('cat.jpg')
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(image_gray, 127, 255, 0)

plt.imshow(cv2.cvtColor(thresh, cv2.COLOR_GRAY2RGB))
plt.show()

contours = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)[0]
image = cv2.drawContours(image, contours, -1, (0, 255, 0), 4)

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```
